```python
import cv2
from google.colab.patches import cv2_imshow
import numpy as np

# Read in image in grayscale
img = cv2.imread('/content/image2.jpg', cv2.IMREAD_GRAYSCALE)

# Gradient operator (Sobel)
sobel_x = cv2.Sobel(img, cv2.CV_64F, 1, 0, ksize=3)
sobel_y = cv2.Sobel(img, cv2.CV_64F, 0, 1, ksize=3)
sobel = np.sqrt(np.square(sobel_x) + np.square(sobel_y))

# Laplacian operator
laplacian = cv2.Laplacian(img, cv2.CV_64F)

# Show original image and edge detection results
cv2_imshow( img)
cv2_imshow(sobel.astype(np.uint8))
cv2_imshow(laplacian.astype(np.uint8))
cv2.waitKey(0)
cv2.destroyAllWindows()
```
